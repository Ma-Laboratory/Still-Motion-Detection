```python
# -*- coding: utf-8 -*-
# =============================================================================
# Complete Code: Logic-Based FMCW Radar Still/Motion Detection (Unsupervised Thresholding)
# Suitable for Google Colab
# Version: Accuracy Calculation based on Ground Truth: 10s Still, 10s Motion, repeated (0-40s)
# MODIFIED to process Range Bins 12-23 and sum variance across ALL chirps and these bins
# =============================================================================

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib # For setting English fonts
import os
import pandas as pd # Needed for fillna

# --- Set English Fonts ---
try:
    # Use a default sans-serif font like Arial, Helvetica, or DejaVu Sans
    # If 'sans-serif' doesn't work, try specific fonts known to be in Colab
    plt.rcParams['font.family'] = 'DejaVu Sans' # Often available in Colab
    plt.rcParams['axes.unicode_minus'] = True       # Ensure minus sign displays correctly
    print("Font set to DejaVu Sans.")
except Exception as e:
    print(f"Failed to set DejaVu Sans font: {e}")
    print("Attempting to set generic sans-serif font.")
    try:
        plt.rcParams['font.family'] = 'sans-serif' # Fallback to generic sans-serif
        plt.rcParams['axes.unicode_minus'] = True
        print("Font set to generic sans-serif.")
    except Exception as e2:
        print(f"Failed to set generic sans-serif font: {e2}")
        print("Using default matplotlib settings, which might not display all characters correctly.")


# --- 1. Load your uploaded data file ---
# IMPORTANT: Make sure your .npy data file is uploaded to your Colab session's /content/ directory
# You need to CHANGE the 'file_name' variable below to match your uploaded file's name.
# For example, if your file is named 'my_radar_data.npy', change the line to:
# file_name = 'my_radar_data.npy'

file_name = 'Lu_lab_still_ 2hand motion_250421_1829.npy' # <--- *** CHANGE THIS TO YOUR UPLOADED FILE NAME ***

file_path = f'/content/{file_name}' # Uploaded files are typically in /content/

print(f"Attempting to load file: {file_path}")
if os.path.exists(file_path):
    try:
        # Load the .npy file
        radar_data_cube = np.load(file_path)
        print("File loaded successfully!")
        print(f"     - Data shape: {radar_data_cube.shape}") # Example shape: (960, 16, 12, 64)
        print(f"     - Data type: {radar_data_cube.dtype}")  # Should be complex or complex128
        # Ensure data is complex if it's not already (e.g., loaded as float from incorrect saving)
        if not np.iscomplexobj(radar_data_cube):
             print("Warning: Loaded data is not complex. Attempting to convert to complex.")
             # Assuming it might be loaded as float representing real part, pad with zeros for imag
             if radar_data_cube.ndim == 4:
                 # This conversion assumes the float data is just the real part.
                 # If it's interleaved real/imag, this won't work correctly.
                 radar_data_cube = radar_data_cube + 0j # Convert to complex
                 print("Converted data to complex by assuming real part.")
             else:
                  print("Error: Cannot convert non-complex data of unexpected shape.")
                  raise TypeError("Loaded data is not complex and has unexpected shape for conversion.")


    except Exception as e:
        print(f"Error loading .npy file: {e}")
        raise # Stop execution
else:
    print(f"Error: Could not find the uploaded file '{file_path}'.")
    print("Please ensure you have successfully uploaded the file to the current Colab session's /content/ directory.")
    print(f"And that the 'file_name' variable in the code is set to '{file_name}'.")
    raise FileNotFoundError(f"File not found at {file_path}") # Stop execution

# --- 2. Preprocessing: Extract target I/Q data for specified range bins and ALL chirps ---
# Modify these based on your radar configuration and where your target is expected
# <<<--- CHANGE THIS TO THE DESIRED RANGE OF BIN INDICES --->>>
# Analyze range bins from 12 up to (but not including) 24, which covers 12 to 23.
target_range_bin_indices = np.arange(12, 24) # <--- Analyze bins 12, 13, ..., 23

target_antenna_index = 0  # <--- Modify antenna index (0 to num_antennas_raw-1) here if needed

print(f"\nPreprocessing parameters:")
print(f"     - Target range bin indices: {target_range_bin_indices}")
print(f"     - Used virtual antenna index: {target_antenna_index}")
print(f"     - Processing data from ALL chirps.")


# Check if the indices are valid
num_frames_raw, num_chirps_raw, num_antennas_raw, num_range_bins_raw = radar_data_cube.shape
if not np.all((target_range_bin_indices >= 0) & (target_range_bin_indices < num_range_bins_raw)):
    raise ValueError(f"One or more target range bin indices {target_range_bin_indices} are out of range [0, {num_range_bins_raw-1}] (Data shape: {radar_data_cube.shape})")
if not (0 <= target_antenna_index < num_antennas_raw):
    raise ValueError(f"Target antenna index {target_antenna_index} is out of range [0, {num_antennas_raw-1}] (Data shape: {radar_data_cube.shape})")
if len(target_range_bin_indices) == 0:
     raise ValueError("No target range bin indices provided. Please specify at least one index.")


# Extract data for the specific range bins and antenna for ALL chirps
# Shape becomes (num_frames, num_chirps, num_selected_bins)
target_data_slice = radar_data_cube[:, :, target_antenna_index, target_range_bin_indices]
print(f"Shape of target slice after extraction (frames, chirps, selected_bins): {target_data_slice.shape}") # e.g., (960, 16, 12)

# Note: We are NOT averaging chirps or bins at this stage. The data cube slice
# retains the frame, chirp, and selected_bin dimensions.
# I/Q separation will implicitly happen when calculating variance in the windowing loop.
# Global i_signal and q_signal are not used in this multi-dimensional approach.


# --- 3. Set analysis parameters ---
# Frame period in milliseconds. This is crucial for converting frame counts to time.
# *** Ensure this matches your radar configuration's frame period! ***
FRAME_PERIOD = 62.5 # milliseconds (Example value: 62.5 ms corresponds to 16 Hz frame rate)
FRAME_RATE = 1000 / FRAME_PERIOD # Hz -> e.g., 16 Hz
fs = FRAME_RATE
print(f"\nUsing sampling rate/frame rate fs: {fs} Hz (derived from FRAME_PERIOD = {FRAME_PERIOD} ms)")

# Generate timestamps t based on the number of frames available
num_frames = target_data_slice.shape[0] # Use length of the extracted data
if num_frames == 0:
    raise ValueError("Processed data slice is empty, cannot continue.")
t = np.arange(num_frames) / fs
print(f"Signal contains {num_frames} frames of data.")
total_duration = t[-1] if num_frames > 0 else 0
print(f"Generated timestamp range: 0 to {total_duration:.2f} seconds.") # e.g., close to 60 seconds

# Logical analysis window parameters
window_size_sec = 1.0     # Sliding window size (seconds) - can be adjusted
step_size_sec = 0.1     # Step size (seconds) - can be adjusted
window_size_samples = int(window_size_sec * fs)
step_size_samples = max(1, int(step_size_sec * fs)) # Ensure step is at least 1 sample

# Basic validation for window/step vs data length
if window_size_samples > num_frames:
    print(f"\nError: Window size ({window_size_samples} samples) is larger than the total number of frames ({num_frames}).")
    print("Please reduce window_size_sec or check your data/FRAME_PERIOD.")
    raise ValueError("Window size exceeds data length.")
if step_size_samples == 0:
     print("Error: Calculated step size is zero. Please check step_size_sec and FRAME_PERIOD.")
     raise ValueError("Calculated step size is zero.")

print(f"Sliding window size: {window_size_sec:.2f} s ({window_size_samples} samples)")
print(f"Sliding window step: {step_size_sec:.2f} s ({step_size_samples} samples)")


# --- 4. Calculate activity index (sliding window variance sum across selected bins AND chirps) ---
activity_index = []
window_times = []

print("\nCalculating activity index (sum of variances across selected bins AND chirps)...")
if num_frames < window_size_samples:
     print(f"Warning: Number of data points ({num_frames}) is less than the window size ({window_size_samples}). Skipping activity index calculation.")
     # activity_index and window_times remain empty lists

# Ensure the loop runs correctly even if step_size_samples is large
num_windows = 0
# Iterate through the frames using the step size
for i in range(0, num_frames - window_size_samples + 1, step_size_samples):
    # Extract data window retaining chirp and bin dimensions
    # Shape: (window_size_samples, num_chirps, num_selected_bins)
    window_data = target_data_slice[i : i + window_size_samples, :, :]

    # Calculate variance of I and Q components over the TIME SAMPLES (axis=0)
    # for *each* chirp and *each* selected bin individually.
    # Result shape: (num_chirps, num_selected_bins)
    var_i_per_chirp_per_bin = np.var(np.real(window_data), axis=0)
    var_q_per_chirp_per_bin = np.var(np.imag(window_data), axis=0)

    # Sum the variances across ALL chirps (axis=0) AND ALL selected bins (axis=1)
    # to get a single activity value for this window.
    current_activity = np.sum(var_i_per_chirp_per_bin + var_q_per_chirp_per_bin)

    # Use max to avoid log10(0) or log10(negative) if variance is extremely small
    # Add a small epsilon to handle exactly zero variance gracefully for log scale plotting later
    activity_index.append(max(current_activity, 1e-18)) # Use a very small positive epsilon

    # Calculate the center time of the window
    window_start_time = t[i] # t is the original full time axis
    window_center_time = window_start_time + window_size_sec / 2
    window_times.append(window_center_time)
    num_windows += 1

if num_windows > 0:
    activity_index = np.array(activity_index)
    window_times = np.array(window_times)
    print(f"Activity index calculation complete. Calculated {len(activity_index)} activity values.")
elif num_frames >= window_size_samples:
     print(f"Warning: Activity index calculation loop resulted in 0 windows. Check step_size_samples ({step_size_samples}) relative to data length and window size.")
     print(f"num_frames: {num_frames}, window_size_samples: {window_size_samples}")
     activity_index = np.array([])
     window_times = np.array([])
else:
    print("Skipping further processing as activity index could not be calculated due to insufficient data.")
    activity_index = np.array([])
    window_times = np.array([])


# --- 5. Unsupervised threshold determination ---
suggested_threshold = np.nan # Initialize
use_suggested_threshold = True # Flag to track if we should use the suggestion

if len(activity_index) > 0: # Proceed only if activity index was calculated
    # 5.1 Plot histogram
    plt.figure(figsize=(10, 6))
    # Filter out potential zero values before log10 for plotting, use a small epsilon
    log_activity_index_for_hist = np.log10(activity_index[activity_index > 1e-18]) # Use the same lower bound
    if len(log_activity_index_for_hist) > 0:
        hist_values, bin_edges, _ = plt.hist(log_activity_index_for_hist, bins=50, density=True, alpha=0.7, color='blue')
        # Update title to reflect multiple bins and chirps
        plt.title('Histogram of Activity Index (log10(Sum of Var_I + Var_Q across selected bins AND chirps))')
        plt.xlabel('log10(Activity Index)')
        plt.ylabel('Probability Density')
        plt.grid(True)
        print("\nPlease observe the histogram to evaluate the suggested threshold.")

        # 5.2 Suggest threshold using percentile
        percentile = 85 # Can be adjusted. This percentile is applied to the *original* activity index values.
        # Calculate percentile on the original activity index, not the log version
        suggested_threshold = np.percentile(activity_index, percentile)

        # Fallback if suggested threshold is too small or invalid
        if not np.isfinite(suggested_threshold) or suggested_threshold < np.percentile(activity_index, 1): # Check if suggested is below a very low percentile
            print(f"\nWarning: Suggested threshold ({suggested_threshold:.2e}) based on {percentile}th percentile is invalid or potentially too low.")
            median_activity = np.median(activity_index[activity_index > 1e-18]) if np.any(activity_index > 1e-18) else 1e-12
            if median_activity > 1e-6:
                 suggested_threshold = median_activity * 5
            else:
                 suggested_threshold = max(median_activity * 100, 1e-4)

            print(f"     - Resetting suggested threshold based on median/fallback logic: {suggested_threshold:.2e}")
            use_suggested_threshold = False # Indicate suggestion might be unreliable

        # Show suggested threshold on the plot
        log_suggested_threshold = np.log10(max(suggested_threshold, 1e-18))
        print(f"\nAutomatically suggested threshold: {suggested_threshold:.2e} (log10 ≈ {log_suggested_threshold:.2f})")
        plt.axvline(log_suggested_threshold, color='red', linestyle='--',
                    label=f'Automatic Suggested Threshold ({percentile}th Percentile)\nlog10 ≈ {log_suggested_threshold:.2f}')
        plt.legend()
        plt.show() # Show the histogram
    else:
        print("Warning: All valid activity indices are too close to zero. Cannot generate histogram or reliable threshold.")
        suggested_threshold = 1e-5 # Default fallback
        use_suggested_threshold = False # Suggestion is just a guess
        print(f"Setting a default low suggested threshold: {suggested_threshold:.2e}")

else: # If activity_index is empty
    print("Warning: Activity index is empty. Cannot determine threshold.")
    suggested_threshold = 1.0 # Assign a default fallback
    use_suggested_threshold = False # Suggestion is just a guess

# 5.3 Determine final threshold
if use_suggested_threshold and np.isfinite(suggested_threshold) and suggested_threshold > 1e-12:
    motion_threshold = suggested_threshold
    print(f"-> Initially using automatically suggested threshold: {motion_threshold:.2e}")
else:
    motion_threshold = 1.0 # Default fallback value
    print(f"-> Initially using fallback threshold: {motion_threshold:.2e}")
    print("     (Automatic suggestion was invalid, unreliable, or data was insufficient)")


# #############################################################################
# <<<--- !!! MANUAL THRESHOLD ADJUSTMENT SECTION - 您可以在这里手动修改阈值 !!! --->>>
#
# Explanation:
# 1. If you want to force a specific threshold value, uncomment ONE of the lines below (remove the #).
# 2. Change the value after the '=' sign to your desired threshold.
# 3. Ensure only ONE `motion_threshold = ...` line in this section is uncommented.
# 4. If you set a value here, it will OVERRIDE the automatically suggested or fallback value above.
# 5. Review the histogram and the result plot below to find a suitable manual threshold if needed.
#    NOTE: The scale of the activity index when summing variances across multiple bins
#    AND chirps will be MUCH larger than when using a single bin or averaging chirps.
#    You WILL likely need to adjust the manual threshold significantly.
#    Look at the y-axis scale of the activity index plot and the histogram's x-axis
#    (log scale) to guide your manual selection.
#
# Example: Thresholds here might be in the range of 1e7 to 1e9 depending on data scale!
# motion_threshold = 50000000.0  # Example manual threshold 5e7
# motion_threshold = 1.0e8      # Another example manual threshold 1e8

#motion_threshold = 3e6      # <--- Uncomment and modify this line if needed
motion_threshold = 1.4e9 # <--- Uncomment and modify this line if needed
#
# #############################################################################


# Final check and print the threshold value ACTUALLY being used
print(f"\nFINAL threshold to be used for classification (motion_threshold): {motion_threshold:.2e}")
print("!!! Please REVIEW this value. If it's not appropriate based on the histogram and results,")
print("     modify the 'motion_threshold = ...' line in the 'MANUAL THRESHOLD ADJUSTMENT SECTION' above and RE-RUN this cell !!!")
# Update print to show selected bins and processing all chirps
print(f"     (Using range bin indices: {target_range_bin_indices}, Antenna: {target_antenna_index}, Processing ALL chirps)")


# --- 6. Classification ---
# Proceed only if activity index was calculated
if len(activity_index) > 0:
    classification = np.zeros_like(activity_index, dtype=int) # 0: Still, 1: Motion
    classification[activity_index > motion_threshold] = 1
    print("Classification complete based on activity index.")
else:
    print("Skipping classification as activity index is empty.")
    classification = np.array([]) # Ensure it's an empty array


# --- 7. Visualize results ---
print("Generating result plots...")
# Adjust title to include selected range bins and mention processing all chirps
title_str = f'File: {file_name}\nTarget Bins: {target_range_bin_indices}, Antenna: {target_antenna_index}, All Chirps, Final Threshold: {motion_threshold:.2e}'
fig, axes = plt.subplots(3, 1, figsize=(12, 10), sharex=True)
fig.suptitle(title_str, fontsize=10)


# Plot 1: Averaged I/Q signals across selected range bins AND chirps
# Show the average I/Q across all selected bins and chirps for a representative view
if 'target_data_slice' in locals() and len(target_data_slice) > 0:
    # Calculate average I/Q across chirps (axis=1) and bins (axis=2)
    avg_frame_complex_signal = np.mean(target_data_slice, axis=(1, 2)) # Average over chirps AND selected bins
    avg_i_signal = np.real(avg_frame_complex_signal)
    avg_q_signal = np.imag(avg_frame_complex_signal)

    axes[0].plot(t, avg_i_signal, label='Avg I Signal (Selected Bins & Chirps)', color='red', alpha=0.8, linewidth=0.8)
    axes[0].plot(t, avg_q_signal, label='Avg Q Signal (Selected Bins & Chirps)', color='blue', alpha=0.8, linewidth=0.8)
    axes[0].set_ylabel('Amplitude')
    # Update plot title
    axes[0].set_title(f'Averaged I/Q Signals Across Selected Bins & Chirps (Total Duration: {total_duration:.2f}s)')
    axes[0].legend(fontsize='small')
    axes[0].grid(True, linestyle='--', alpha=0.6)
else:
    axes[0].set_title('No Averaged I/Q Data to Plot')

# Plot 2: Activity index and threshold
if len(activity_index) > 0 and len(window_times) == len(activity_index):
    # Update label to reflect summing across bins and chirps
    axes[1].plot(window_times, activity_index, label='Activity Index (Sum of Var_I + Var_Q across selected bins & chirps)', color='green', linewidth=1.0)
    axes[1].axhline(motion_threshold, color='magenta', linestyle='--', label=f'Final Threshold ({motion_threshold:.2e})')
    try:
        if np.all(activity_index > 1e-18):
              axes[1].set_yscale('log')
              # Update label
              axes[1].set_ylabel('Sum of Variances (Log Scale)')
              axes[1].grid(True, which='major', axis='y', linestyle='-', alpha=0.7)
              axes[1].grid(True, which='minor', axis='y', linestyle=':', alpha=0.4)
        else:
              print("Warning: Activity index contains non-positive values, using linear scale for plot 2.")
              axes[1].set_yscale('linear')
              # Update label
              axes[1].set_ylabel('Sum of Variances (Linear Scale)')
              axes[1].grid(True, linestyle='--', alpha=0.6)

    except ValueError: # Catch potential errors during yscale setting
        print("Error setting y-axis scale for plot 2, using linear scale.")
        axes[1].set_yscale('linear')
        axes[1].set_ylabel('Sum of Variances (Linear Scale)')
        axes[1].grid(True, linestyle='--', alpha=0.6)

    # Update plot title
    axes[1].set_title('Activity Index vs. Selected Threshold')
    axes[1].legend(fontsize='small')
    axes[1].grid(True, which='major', axis='x', linestyle='--', alpha=0.6) # Grid for x-axis
    if num_frames > 0:
      axes[1].set_xlim(t[0], t[-1])

else:
    axes[1].set_title('No Activity Index Data to Plot')
    if num_frames > 0:
        axes[1].set_xlim(t[0], t[-1])


# Plot 3: Classification results mapped back to original time axis
# This part works with the single activity index derived from multiple bins and chirps.
classification_full = np.full(num_frames, np.nan) # Initialize with NaNs

if len(classification) > 0 and len(window_times) == len(classification) and num_frames > 0:
    # Map windowed classification back to the full time series
    activity_index_original_frame_indices = np.arange(0, num_frames - window_size_samples + 1, step_size_samples)

    classification_map = np.full(num_frames, np.nan) # Use NaN to indicate unassigned

    # Iterate through the calculated classification values
    for k in range(len(classification)):
        # Original start index of the window that produced this classification value
        original_start_idx = activity_index_original_frame_indices[k]
        original_end_idx = original_start_idx + window_size_samples

        # Assign the classification value to the original indices covered by this window
        classification_map[original_start_idx:min(original_end_idx, num_frames)] = classification[k]

    # Fill any remaining NaN values
    classification_map = np.array(classification_map)
    if classification_map.size > 0 and not np.all(np.isnan(classification_map)):
        classification_series = pd.Series(classification_map)
        classification_series = classification_series.fillna(method='ffill') # Forward fill
        classification_series = classification_series.fillna(method='bfill') # Backward fill (for NaNs at the start)
        classification_full = classification_series.values.astype(int)
    else:
         print("Warning: classification_map is empty or all NaN after mapping. Cannot fill NaNs.")
         classification_full = np.array(classification_map)

    # Actual plotting using fill_between
    if not np.all(np.isnan(classification_full)):
        axes[2].fill_between(t, 0, 1, where=classification_full==1, color='orange', alpha=0.6, step='post', label='Motion Detected')
        axes[2].fill_between(t, 0, 1, where=classification_full==0, color='cyan', alpha=0.6, step='post', label='Still Detected')
        axes[2].set_yticks([0, 1])
        axes[2].set_yticklabels(['Still (0)', 'Motion (1)'])
        axes[2].set_title('Detected State (Mapped to Full Time Axis)')
    else:
         axes[2].set_title('Detected State (No Valid Classification to Plot)')

else: # No classification data
      axes[2].set_title('No Classification Data to Plot')
      if num_frames > 0:
          classification_full = np.full(num_frames, np.nan)
      else:
           classification_full = np.array([])


# Common settings for Plot 3
axes[2].set_xlabel('Time (seconds)')
axes[2].set_ylabel('State')
axes[2].set_ylim(-0.1, 1.1) # Slightly extend limits for visibility
axes[2].legend(fontsize='small')
axes[2].grid(True, linestyle='--', alpha=0.6)
if num_frames > 0:
    axes[2].set_xlim(t[0], t[-1])


# --- 8. Calculate Accuracy based on User-Provided Ground Truth Timing ---
# <<<--- MODIFIED CODE BLOCK START for 10s Still / 10s Motion Pattern --->>>
print("\nCalculating accuracy based on NEW ground truth timing pattern:")
print("Pattern: 0-10s Still, 10-20s Motion, 20-30s Still, 30-40s Motion.")

# Initialize ground truth labels to Still (0) for the full duration of the signal
ground_truth_labels = np.zeros(num_frames, dtype=int)

# Define the time points for state changes for the ground truth pattern (up to 40s)
gt_time_0s = 0.0
gt_time_10s = 10.0
gt_time_20s = 20.0
gt_time_30s = 30.0
gt_time_40s = 40.0
pattern_duration_sec = gt_time_40s # The pattern is defined up to 40 seconds

print(f"Ground truth pattern defined up to {pattern_duration_sec:.2f} seconds.")
if total_duration < pattern_duration_sec:
      print(f"Warning: Signal duration ({total_duration:.2f}s) is shorter than the defined ground truth pattern duration ({pattern_duration_sec:.2f}s).")
      print(f"         Accuracy will be calculated based on available signal length up to {total_duration:.2f}s.")
      comparison_duration_sec = total_duration
else:
      comparison_duration_sec = pattern_duration_sec


# Convert ground truth time points to frame indices, clamping to signal length
gt_idx_0s = 0
gt_idx_10s = min(int(gt_time_10s * fs), num_frames)
gt_idx_20s = min(int(gt_time_20s * fs), num_frames)
gt_idx_30s = min(int(gt_time_30s * fs), num_frames)
gt_idx_40s = min(int(gt_time_40s * fs), num_frames)

print(f"Ground truth indices (based on signal length and pattern): 0s={gt_idx_0s}, 10s={gt_idx_10s}, 20s={gt_idx_20s}, 30s={gt_idx_30s}, 40s={gt_idx_40s}")


# Apply the ground truth pattern
if gt_idx_10s < gt_idx_20s:
    ground_truth_labels[gt_idx_10s:gt_idx_20s] = 1

if gt_idx_30s < gt_idx_40s:
    ground_truth_labels[gt_idx_30s:gt_idx_40s] = 1


# Ensure classification_full is valid and has the correct length for comparison
accuracy = np.nan # Default to NaN
if 'classification_full' in locals() and isinstance(classification_full, np.ndarray) and len(classification_full) == num_frames and num_frames > 0 and not np.all(np.isnan(classification_full)):
    comparison_length_frames = min(num_frames, int(comparison_duration_sec * fs))
    if comparison_length_frames == 0 and num_frames > 0:
         comparison_length_frames = min(1, num_frames)

    if comparison_length_frames > 0:
      valid_comparison_indices = ~np.isnan(classification_full[:comparison_length_frames])

      if np.any(valid_comparison_indices):
           correct_predictions = (classification_full[:comparison_length_frames][valid_comparison_indices] == ground_truth_labels[:comparison_length_frames][valid_comparison_indices])
           accuracy = np.mean(correct_predictions) * 100
           num_compared_frames = np.sum(valid_comparison_indices)
           print(f"Accuracy compared to specified timing pattern: {accuracy:.2f}% (calculated over {num_compared_frames} frames within the first {comparison_duration_sec:.2f} seconds)")
      else:
          print("Cannot calculate accuracy: Classification results within the comparison duration are all invalid (NaN).")
          accuracy = np.nan

    else:
      print("Cannot calculate accuracy: No valid comparison duration or zero frames to compare.")


    # Optional: Add the ground truth to the plot for visual comparison
    if num_frames > 0:
      gt_plot_signal = np.zeros_like(t, dtype=int)
      if gt_idx_10s < num_frames:
          gt_plot_signal[gt_idx_10s:min(gt_idx_20s, num_frames)] = 1
      if gt_idx_30s < num_frames:
          gt_plot_signal[gt_idx_30s:min(gt_idx_40s, num_frames)] = 1

      axes[2].step(t, gt_plot_signal + 0.05, 'k--', alpha=0.7, where='post', label='Ground Truth Timing (0-40s Pattern)')

      handles, labels = axes[2].get_legend_handles_labels()
      unique_labels = []
      unique_handles = []
      for h, l in zip(handles, labels):
          if l not in unique_labels:
              unique_labels.append(l)
              unique_handles.append(h)
      axes[2].legend(unique_handles, unique_labels, fontsize='small')

else:
    print("Error: Could not calculate accuracy. Classification results might not be available or have incorrect length/NaN values, or signal is empty.")
    if num_frames == 0:
        ground_truth_labels = np.array([])


# <<<--- MODIFIED CODE BLOCK END --->>>

# Adjust layout and show the plots
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Final Instructions
print("\nCode execution complete. Please carefully examine the plots:")
print("1. Histogram: Helps evaluate the automatic threshold suggestion.")
print("2. Result Plot:")
print("     - Top plot: Averaged I/Q signals across selected range bins AND chirps.") # Updated description
print("     - Middle plot: Activity index (Sum of variances across selected bins & chirps) vs. the final threshold.") # Updated description
print("     - Bottom plot: Detected state (Still/Motion) mapped to the full time axis, with Ground Truth overlaid.")
print(f"\nAccuracy calculated against the 0-40s pattern: {accuracy:.2f}% (if valid)")
print("\nReview the threshold value printed above and the plots. If results are not satisfactory, manually adjust the 'motion_threshold' in the dedicated section and re-run the cell.")
```

```python
# -*- coding: utf-8 -*-
# =============================================================================
# Complete Code: Logic-Based FMCW Radar Still/Motion Detection (Unsupervised Thresholding)
# Suitable for Google Colab
# Version: Accuracy Calculation based on Ground Truth: 10s Still, 10s Motion, repeated (0-40s)
# =============================================================================

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib # For setting English fonts
import os

# --- Set English Fonts ---
try:
    # Use a default sans-serif font like Arial, Helvetica, or DejaVu Sans
    # If 'sans-serif' doesn't work, try specific fonts known to be in Colab
    plt.rcParams['font.family'] = 'DejaVu Sans' # Often available in Colab
    plt.rcParams['axes.unicode_minus'] = True      # Ensure minus sign displays correctly
    print("Font set to DejaVu Sans.")
except Exception as e:
    print(f"Failed to set DejaVu Sans font: {e}")
    print("Attempting to set generic sans-serif font.")
    try:
        plt.rcParams['font.family'] = 'sans-serif' # Fallback to generic sans-serif
        plt.rcParams['axes.unicode_minus'] = True
        print("Font set to generic sans-serif.")
    except Exception as e2:
         print(f"Failed to set generic sans-serif font: {e2}")
         print("Using default matplotlib settings, which might not display all characters correctly.")


# --- 1. Load your uploaded data file ---
# IMPORTANT: Make sure your .npy data file is uploaded to your Colab session's /content/ directory
# You need to CHANGE the 'file_name' variable below to match your uploaded file's name.
# For example, if your file is named 'my_radar_data.npy', change the line to:
# file_name = 'my_radar_data.npy'

file_name = 'Lu_lab_still_ 2hand motion_250421_1829.npy' # <--- *** CHANGE THIS TO YOUR UPLOADED FILE NAME ***

file_path = f'/content/{file_name}' # Uploaded files are typically in /content/

print(f"Attempting to load file: {file_path}")
if os.path.exists(file_path):
    try:
        # Load the .npy file
        radar_data_cube = np.load(file_path)
        print("File loaded successfully!")
        print(f"    - Data shape: {radar_data_cube.shape}") # Example shape: (960, 16, 12, 64)
        print(f"    - Data type: {radar_data_cube.dtype}")  # Should be complex or complex128
        # Ensure data is complex if it's not already (e.g., loaded as float from incorrect saving)
        if not np.iscomplexobj(radar_data_cube):
             print("Warning: Loaded data is not complex. Attempting to convert to complex.")
             # Assuming it might be loaded as float representing real part, pad with zeros for imag
             if radar_data_cube.ndim == 4:
                 radar_data_cube = radar_data_cube + 0j # Convert to complex
             else:
                 print("Error: Cannot convert non-complex data of unexpected shape.")
                 raise TypeError("Loaded data is not complex and has unexpected shape for conversion.")


    except Exception as e:
        print(f"Error loading .npy file: {e}")
        raise # Stop execution
else:
    print(f"Error: Could not find the uploaded file '{file_path}'.")
    print("Please ensure you have successfully uploaded the file to the current Colab session's /content/ directory.")
    print(f"And that the 'file_name' variable in the code is set to '{file_name}'.")
    raise FileNotFoundError(f"File not found at {file_path}") # Stop execution

# --- 2. Preprocessing: Extract target I/Q time series from the data cube ---
# Modify these indices based on your radar configuration and where your target is expected
target_range_bin_index = 16 # <--- Modify target range bin index (0 to num_range_bins_raw-1) here if needed
target_antenna_index = 0  # <--- Modify antenna index (0 to num_antennas_raw-1) here if needed

print(f"\nPreprocessing parameters:")
print(f"    - Target range bin index: {target_range_bin_index}")
print(f"    - Used virtual antenna index: {target_antenna_index}")

# Check if the indices are valid
num_frames_raw, num_chirps_raw, num_antennas_raw, num_range_bins_raw = radar_data_cube.shape
if not (0 <= target_range_bin_index < num_range_bins_raw):
    raise ValueError(f"Target range bin index {target_range_bin_index} is out of range [0, {num_range_bins_raw-1}] (Data shape: {radar_data_cube.shape})")
if not (0 <= target_antenna_index < num_antennas_raw):
    raise ValueError(f"Target antenna index {target_antenna_index} is out of range [0, {num_antennas_raw-1}] (Data shape: {radar_data_cube.shape})")

# Extract data for the specific range bin and antenna, shape becomes (num_frames, num_chirps)
target_data_slice = radar_data_cube[:, :, target_antenna_index, target_range_bin_index]
print(f"Shape of target slice after extraction: {target_data_slice.shape}") # e.g., (960, 16)

# Average the Chirp data within each frame to get one complex value per frame
# Ensure the data is complex before averaging
if not np.iscomplexobj(target_data_slice):
     print("Warning: Target data slice is not complex before averaging chirps. This is unexpected.")
     # Attempt conversion, though this should ideally not happen if load was correct
     target_data_slice = target_data_slice.astype(complex)


frame_complex_signal = np.mean(target_data_slice, axis=1)
print(f"Shape of data after averaging Chirps: {frame_complex_signal.shape}") # e.g., (960,)

# Separate I and Q signals
i_signal = np.real(frame_complex_signal)
q_signal = np.imag(frame_complex_signal)
print("Successfully separated I and Q signal time series.")

# --- 3. Set analysis parameters ---
# Frame period in milliseconds. This is crucial for converting frame counts to time.
# *** Ensure this matches your radar configuration's frame period! ***
FRAME_PERIOD = 62.5 # milliseconds (Example value: 62.5 ms corresponds to 16 Hz frame rate)
FRAME_RATE = 1000 / FRAME_PERIOD # Hz -> e.g., 16 Hz
fs = FRAME_RATE
print(f"\nUsing sampling rate/frame rate fs: {fs} Hz (derived from FRAME_PERIOD = {FRAME_PERIOD} ms)")

# Generate timestamps t
num_frames = len(i_signal) # e.g., 960
if num_frames == 0:
    raise ValueError("Processed signal is empty, cannot continue.")
t = np.arange(num_frames) / fs
print(f"Signal contains {num_frames} frames of data.")
total_duration = t[-1] if num_frames > 0 else 0
print(f"Generated timestamp range: 0 to {total_duration:.2f} seconds.") # e.g., close to 60 seconds

# Logical analysis window parameters
window_size_sec = 1.0   # Sliding window size (seconds) - can be adjusted
step_size_sec = 0.1     # Step size (seconds) - can be adjusted
window_size_samples = int(window_size_sec * fs)
step_size_samples = max(1, int(step_size_sec * fs)) # Ensure step is at least 1 sample

# Basic validation for window/step vs data length
if window_size_samples > num_frames:
    print(f"\nError: Window size ({window_size_samples} samples) is larger than the total number of frames ({num_frames}).")
    print("Please reduce window_size_sec or check your data/FRAME_PERIOD.")
    raise ValueError("Window size exceeds data length.")
if step_size_samples == 0:
     print("Error: Calculated step size is zero. Please check step_size_sec and FRAME_PERIOD.")
     raise ValueError("Calculated step size is zero.")

print(f"Sliding window size: {window_size_sec:.2f} s ({window_size_samples} samples)")
print(f"Sliding window step: {step_size_sec:.2f} s ({step_size_samples} samples)")


# --- 4. Calculate activity index (sliding window variance sum) ---
activity_index = []
window_times = []

print("\nCalculating activity index...")
if num_frames < window_size_samples:
     # This case should be caught by the validation above, but adding a print here just in case
     print(f"Warning: Number of data points ({num_frames}) is less than the window size ({window_size_samples}). Skipping activity index calculation.")
     # activity_index and window_times remain empty lists

# Ensure the loop runs correctly even if step_size_samples is large
# The loop condition ensures 'i + window_size_samples' does not exceed 'num_frames'
num_windows = 0
# The range function is correct: it stops when i + window_size_samples > num_frames
for i in range(0, num_frames - window_size_samples + 1, step_size_samples):
    i_window = i_signal[i : i + window_size_samples]
    q_window = q_signal[i : i + window_size_samples]

    # Variance is calculated for each window
    var_i = np.var(i_window)
    var_q = np.var(q_window)
    current_activity = var_i + var_q
    # Use max to avoid log10(0) or log10(negative) if variance is extremely small
    # Add a small epsilon to handle exactly zero variance gracefully for log scale plotting later
    activity_index.append(max(current_activity, 1e-18)) # Use a very small positive epsilon

    # Calculate the center time of the window
    # The time corresponds to the START of the window + half window duration
    window_start_time = t[i]
    window_center_time = window_start_time + window_size_sec / 2
    window_times.append(window_center_time)
    num_windows += 1

if num_windows > 0:
    activity_index = np.array(activity_index)
    window_times = np.array(window_times)
    print(f"Activity index calculation complete. Calculated {len(activity_index)} activity values.")
elif num_frames >= window_size_samples:
    # This means the loop range was valid but the loop didn't execute, implying a large step size
    print(f"Warning: Activity index calculation loop resulted in 0 windows. Check step_size_samples ({step_size_samples}) relative to data length and window size.")
    print(f"num_frames: {num_frames}, window_size_samples: {window_size_samples}")
    # Keep activity_index and window_times as empty arrays
    activity_index = np.array([])
    window_times = np.array([])
else:
    print("Skipping further processing as activity index could not be calculated due to insufficient data.")
    activity_index = np.array([])
    window_times = np.array([])


# --- 5. Unsupervised threshold determination ---
suggested_threshold = np.nan # Initialize
use_suggested_threshold = True # Flag to track if we should use the suggestion

if len(activity_index) > 0: # Proceed only if activity index was calculated
    # 5.1 Plot histogram
    plt.figure(figsize=(10, 6))
    # Filter out potential zero values before log10 for plotting, use a small epsilon
    log_activity_index_for_hist = np.log10(activity_index[activity_index > 1e-18]) # Use the same lower bound
    if len(log_activity_index_for_hist) > 0:
        hist_values, bin_edges, _ = plt.hist(log_activity_index_for_hist, bins=50, density=True, alpha=0.7, color='blue')
        plt.title('Histogram of Activity Index (log10(Var_I + Var_Q))')
        plt.xlabel('log10(Activity Index)')
        plt.ylabel('Probability Density')
        plt.grid(True)
        print("\nPlease observe the histogram to evaluate the suggested threshold.")

        # 5.2 Suggest threshold using percentile
        percentile = 85 # Can be adjusted. This percentile is applied to the *original* activity index values.
        # Calculate percentile on the original activity index, not the log version
        suggested_threshold = np.percentile(activity_index, percentile)

        # Fallback if suggested threshold is too small or invalid
        # A threshold near zero means almost any tiny variation is considered motion, which is likely wrong.
        # A threshold that is NaN suggests calculation issues.
        if not np.isfinite(suggested_threshold) or suggested_threshold < np.percentile(activity_index, 1): # Check if suggested is below a very low percentile
            print(f"\nWarning: Suggested threshold ({suggested_threshold:.2e}) based on {percentile}th percentile is invalid or potentially too low.")
            # Use a multiplier of the median as a fallback basis
            median_activity = np.median(activity_index[activity_index > 1e-18]) if np.any(activity_index > 1e-18) else 1e-12 # Use a small default if median is zero
            # Adjust multiplier based on expected data scale - this is heuristic
            if median_activity > 1e-6: # If median is reasonably large, use a smaller multiplier
                 suggested_threshold = median_activity * 5
            else: # If median is very small, use a larger multiplier or a fixed small value
                 suggested_threshold = max(median_activity * 100, 1e-4) # Ensure it's not vanishingly small

            print(f"    - Resetting suggested threshold based on median/fallback logic: {suggested_threshold:.2e}")
            use_suggested_threshold = False # Indicate suggestion might be unreliable

        # Show suggested threshold on the plot
        # Use max with epsilon for log plotting the threshold itself
        log_suggested_threshold = np.log10(max(suggested_threshold, 1e-18))
        print(f"\nAutomatically suggested threshold: {suggested_threshold:.2e} (log10 ≈ {log_suggested_threshold:.2f})")
        plt.axvline(log_suggested_threshold, color='red', linestyle='--',
                    label=f'Automatic Suggested Threshold ({percentile}th Percentile)\nlog10 ≈ {log_suggested_threshold:.2f}')
        plt.legend()
        plt.show() # Show the histogram
    else:
        print("Warning: All valid activity indices are too close to zero. Cannot generate histogram or reliable threshold.")
        # Set a default low threshold if all activity is near zero
        suggested_threshold = 1e-5 # Default fallback
        use_suggested_threshold = False # Suggestion is just a guess
        print(f"Setting a default low suggested threshold: {suggested_threshold:.2e}")

else: # If activity_index is empty
    print("Warning: Activity index is empty. Cannot determine threshold.")
    suggested_threshold = 1.0 # Assign a default fallback that will likely result in 'Still' classification
    use_suggested_threshold = False # Suggestion is just a guess

# 5.3 Determine final threshold
# Initially, set motion_threshold based on the suggestion or fallback
if use_suggested_threshold and np.isfinite(suggested_threshold) and suggested_threshold > 1e-12: # Ensure valid and not extremely small
    motion_threshold = suggested_threshold
    print(f"-> Initially using automatically suggested threshold: {motion_threshold:.2e}")
else:
    # Handle cases where suggestion failed, was invalid, or not recommended
    motion_threshold = 1.0 # Default fallback value
    print(f"-> Initially using fallback threshold: {motion_threshold:.2e}")
    print("     (Automatic suggestion was invalid, unreliable, or data was insufficient)")


# #############################################################################
# <<<--- !!! MANUAL THRESHOLD ADJUSTMENT SECTION - 您可以在这里手动修改阈值 !!! --->>>
#
# Explanation:
# 1. If you want to force a specific threshold value, uncomment ONE of the lines below (remove the #).
# 2. Change the value after the '=' sign to your desired threshold.
# 3. Ensure only ONE `motion_threshold = ...` line in this section is uncommented.
# 4. If you set a value here, it will OVERRIDE the automatically suggested or fallback value above.
# 5. Review the histogram and the result plot below to find a suitable manual threshold if needed.
#
# Example:
# motion_threshold = 1000000.0  # Example manual threshold 1e6


















motion_threshold = 3e6      # <--- Uncomment and modify this line if needed (e.g., 2,500,000)
# motion_threshold = 5.0e5      # <--- Uncomment and modify this line if needed (e.g., 500,000)
#
# #############################################################################


# Final check and print the threshold value ACTUALLY being used
print(f"\nFINAL threshold to be used for classification (motion_threshold): {motion_threshold:.2e}")
print("!!! Please REVIEW this value. If it's not appropriate based on the histogram and results,")
print("     modify the 'motion_threshold = ...' line in the 'MANUAL THRESHOLD ADJUSTMENT SECTION' above and RE-RUN this cell !!!")


# --- 6. Classification ---
# Proceed only if activity index was calculated
if len(activity_index) > 0:
    classification = np.zeros_like(activity_index, dtype=int) # 0: Still, 1: Motion
    classification[activity_index > motion_threshold] = 1
    print("Classification complete based on activity index.")
else:
    print("Skipping classification as activity index is empty.")
    classification = np.array([]) # Ensure it's an empty array


# --- 7. Visualize results ---
print("Generating result plots...")
fig, axes = plt.subplots(3, 1, figsize=(12, 10), sharex=True)
# Add plot title here for context even if some plots might be empty
fig.suptitle(f'File: {file_name}\nTarget Bin: {target_range_bin_index}, Antenna: {target_antenna_index}, Final Threshold: {motion_threshold:.2e}', fontsize=10)

# Plot 1: Extracted target I/Q signals
if num_frames > 0:
    axes[0].plot(t, i_signal, label='Extracted I Signal', color='red', alpha=0.8, linewidth=0.8)
    axes[0].plot(t, q_signal, label='Extracted Q Signal', color='blue', alpha=0.8, linewidth=0.8)
    axes[0].set_ylabel('Amplitude')
    axes[0].set_title(f'Extracted I/Q Signals (Total Duration: {total_duration:.2f}s)')
    axes[0].legend(fontsize='small')
    axes[0].grid(True, linestyle='--', alpha=0.6)
else:
    axes[0].set_title('No I/Q Signal Data to Plot')

# Plot 2: Activity index and threshold
if len(activity_index) > 0 and len(window_times) == len(activity_index):
    axes[1].plot(window_times, activity_index, label='Activity Index (Var_I + Var_Q)', color='green', linewidth=1.0)
    axes[1].axhline(motion_threshold, color='magenta', linestyle='--', label=f'Final Threshold ({motion_threshold:.2e})')
    try:
        # Attempt log scale, fall back to linear if values are zero or negative or fail log scale
        # Check for positive values before setting log scale
        if np.all(activity_index > 1e-18): # Use the same epsilon as before
             axes[1].set_yscale('log')
             axes[1].set_ylabel('Sum of Variances (Log Scale)')
             # Add minor grids for log scale
             axes[1].grid(True, which='major', axis='y', linestyle='-', alpha=0.7)
             axes[1].grid(True, which='minor', axis='y', linestyle=':', alpha=0.4)
        else:
             print("Warning: Activity index contains non-positive values, using linear scale for plot 2.")
             axes[1].set_yscale('linear')
             axes[1].set_ylabel('Sum of Variances (Linear Scale)')
             axes[1].grid(True, linestyle='--', alpha=0.6)

    except ValueError: # Catch potential errors during yscale setting
        print("Error setting y-axis scale for plot 2, using linear scale.")
        axes[1].set_yscale('linear')
        axes[1].set_ylabel('Sum of Variances (Linear Scale)')
        axes[1].grid(True, linestyle='--', alpha=0.6)

    axes[1].set_title('Activity Index vs. Selected Threshold')
    axes[1].legend(fontsize='small')
    axes[1].grid(True, which='major', axis='x', linestyle='--', alpha=0.6) # Grid for x-axis
    # Set x-limits to match the overall time duration for consistency
    if num_frames > 0:
      axes[1].set_xlim(t[0], t[-1])

else:
    axes[1].set_title('No Activity Index Data to Plot')
    if num_frames > 0:
         axes[1].set_xlim(t[0], t[-1]) # Still set x-limits if possible


# Plot 3: Classification results mapped back to original time axis
classification_full = np.full(num_frames, np.nan) # Initialize with NaNs

if len(classification) > 0 and len(window_times) == len(classification) and num_frames > 0:
    # Map windowed classification back to the full time series
    # Assign the classification result of a window to the time points covered by that window.
    # A simple way is to assign the window's classification to all original frame indices
    # that fall within the window's time span. If windows overlap, the last window's
    # classification for a given time point will overwrite previous ones.

    # Calculate the start and end indices for each window on the original time axis
    window_start_indices = np.maximum(0, (window_times - window_size_sec/2) * fs).astype(int)
    window_end_indices = np.minimum(num_frames, (window_times + window_size_sec/2) * fs).astype(int)

    # Assign classification values
    # Iterate through windows and assign classification to the frames within that window's time bounds
    # This approach can be simplified. The classification result at window_times[i]
    # corresponds to the segment of data from approx window_times[i] - window_size_sec/2
    # to window_times[i] + window_size_sec/2. Since windows are stepped, we can think
    # of the classification[i] applying roughly from the center of window[i-1] to center of window[i],
    # or more simply, applying it to the original frames that contributed most directly
    # to the activity index calculation at window_times[i].
    # A robust way is to interpolate or nearest-neighbor assign classification values.

    # Let's use nearest neighbor assignment from window_times to original frame times t
    # Find the index in window_times closest to each time point in t
    # This can be computationally intensive for large datasets.
    # A simpler approach is to assign the classification[i] value to the time span covered by the i-th window.
    # If windows overlap, the later classification overwrites.

    # Assign the classification value to the frame indices corresponding to the window's center time.
    # This is a simplification, but often sufficient for visualization.
    # More accurately, we should assign it to the original indices [i : i + window_size_samples]
    # used to calculate activity_index[k], where k is the index corresponding to original index i.
    # The range loop for activity index is range(0, num_frames - window_size_samples + 1, step_size_samples)
    # The k-th activity index is calculated from original indices i = k * step_size_samples
    # So, classification[k] corresponds to original indices [k*step_size_samples : k*step_size_samples + window_size_samples]

    # Map classification back to original full frame indices
    classification_map = np.full(num_frames, np.nan) # Use NaN to indicate unassigned
    activity_index_original_frame_indices = np.arange(0, num_frames - window_size_samples + 1, step_size_samples)

    # Iterate through the calculated classification values
    for k in range(len(classification)):
        # Original start index of the window that produced this classification value
        original_start_idx = activity_index_original_frame_indices[k]
        original_end_idx = original_start_idx + window_size_samples

        # Assign the classification value to the original indices covered by this window
        # If windows overlap, later assignments overwrite.
        classification_map[original_start_idx:original_end_idx] = classification[k]

    # Fill any remaining NaN values (e.g., at the very beginning or end not fully covered by a window)
    # Use forward fill, then backward fill
    classification_map = np.array(classification_map) # Ensure it's a numpy array for ffill/bfill
    # Convert to pandas Series for ffill/bfill, then back to numpy
    import pandas as pd
    classification_series = pd.Series(classification_map)
    classification_series = classification_series.fillna(method='ffill') # Forward fill
    classification_series = classification_series.fillna(method='bfill') # Backward fill (for NaNs at the start)
    classification_full = classification_series.values.astype(int)


    # Actual plotting using fill_between
    axes[2].fill_between(t, 0, 1, where=classification_full==1, color='orange', alpha=0.6, step='post', label='Motion Detected')
    axes[2].fill_between(t, 0, 1, where=classification_full==0, color='cyan', alpha=0.6, step='post', label='Still Detected')
    axes[2].set_yticks([0, 1])
    axes[2].set_yticklabels(['Still (0)', 'Motion (1)'])
    axes[2].set_title('Detected State (Mapped to Full Time Axis)')

else: # No classification data
      axes[2].set_title('No Classification Data to Plot')
      # Ensure classification_full exists as an array of zeros or NaNs for accuracy calculation fallback
      # If num_frames > 0 but classification was empty, create a placeholder
      if num_frames > 0:
          classification_full = np.full(num_frames, np.nan) # Use NaN if no classification was possible
      else:
           classification_full = np.array([]) # Empty if no frames


# Common settings for Plot 3
axes[2].set_xlabel('Time (seconds)')
axes[2].set_ylabel('State')
axes[2].set_ylim(-0.1, 1.1) # Slightly extend limits for visibility
axes[2].legend(fontsize='small')
axes[2].grid(True, linestyle='--', alpha=0.6)
if num_frames > 0:
    axes[2].set_xlim(t[0], t[-1])


# --- 8. Calculate Accuracy based on User-Provided Ground Truth Timing ---
# <<<--- MODIFIED CODE BLOCK START for 10s Still / 10s Motion Pattern --->>>
print("\nCalculating accuracy based on NEW ground truth timing pattern:")
print("Pattern: 0-10s Still, 10-20s Motion, 20-30s Still, 30-40s Motion.")

# Initialize ground truth labels to Still (0) for the full duration of the signal
ground_truth_labels = np.zeros(num_frames, dtype=int)

# Define the time points for state changes for the ground truth pattern (up to 40s)
gt_time_0s = 0.0
gt_time_10s = 10.0
gt_time_20s = 20.0
gt_time_30s = 30.0
gt_time_40s = 40.0
pattern_duration_sec = gt_time_40s # The pattern is defined up to 40 seconds

print(f"Ground truth pattern defined up to {pattern_duration_sec:.2f} seconds.")
if total_duration < pattern_duration_sec:
      print(f"Warning: Signal duration ({total_duration:.2f}s) is shorter than the defined ground truth pattern duration ({pattern_duration_sec:.2f}s).")
      print(f"         Accuracy will be calculated based on available signal length up to {total_duration:.2f}s.")
      comparison_duration_sec = total_duration
else:
      comparison_duration_sec = pattern_duration_sec


# Convert ground truth time points to frame indices, clamping to signal length
# Indices mark the START of a segment with the new state
gt_idx_0s = 0 # Always starts at the beginning
gt_idx_10s = min(int(gt_time_10s * fs), num_frames)
gt_idx_20s = min(int(gt_time_20s * fs), num_frames)
gt_idx_30s = min(int(gt_time_30s * fs), num_frames)
gt_idx_40s = min(int(gt_time_40s * fs), num_frames) # Cap at num_frames

print(f"Ground truth indices (based on signal length and pattern): 0s={gt_idx_0s}, 10s={gt_idx_10s}, 20s={gt_idx_20s}, 30s={gt_idx_30s}, 40s={gt_idx_40s}")


# Apply the ground truth pattern to the ground_truth_labels array
# Segment 0-10s: Still (already initialized to 0)

# Segment 10-20s: Motion (1)
if gt_idx_10s < gt_idx_20s: # Ensure the time segment is valid within the signal
    ground_truth_labels[gt_idx_10s:gt_idx_20s] = 1
    # print(f"  GT Motion segment 1 set from index {gt_idx_10s} to {gt_idx_20s-1}")

# Segment 20-30s: Still (already initialized to 0)

# Segment 30-40s: Motion (1)
if gt_idx_30s < gt_idx_40s: # Ensure the time segment is valid within the signal
    ground_truth_labels[gt_idx_30s:gt_idx_40s] = 1
    # print(f"  GT Motion segment 2 set from index {gt_idx_30s} to {gt_idx_40s-1}")

# For times beyond 40s (if signal is longer), ground truth remains Still (0) as per initialization.


# Ensure classification_full is valid and has the correct length for comparison
accuracy = np.nan # Default to NaN
if 'classification_full' in locals() and isinstance(classification_full, np.ndarray) and len(classification_full) == num_frames and num_frames > 0 :
    # Determine the number of frames to use for comparison
    comparison_length_frames = min(num_frames, int(comparison_duration_sec * fs))
    # Ensure comparison_length_frames is at least 1 if possible
    if comparison_length_frames == 0 and num_frames > 0:
        comparison_length_frames = 1 # Compare at least the first frame if data exists

    if comparison_length_frames > 0:
      # Calculate accuracy over the valid comparison range (up to 40s or signal end, whichever is shorter)
      # Note: classification_full might contain NaNs if classification was skipped.
      # We should only compare where classification_full is not NaN.
      # If classification was skipped entirely, classification_full will be all NaNs.
      valid_comparison_indices = ~np.isnan(classification_full[:comparison_length_frames])

      if np.any(valid_comparison_indices):
          # Compare only the valid (non-NaN) parts of the classification
          correct_predictions = (classification_full[:comparison_length_frames][valid_comparison_indices] == ground_truth_labels[:comparison_length_frames][valid_comparison_indices])
          accuracy = np.mean(correct_predictions) * 100 # np.mean of boolean array treats True=1, False=0
          num_compared_frames = np.sum(valid_comparison_indices)
          print(f"Accuracy compared to specified timing pattern: {accuracy:.2f}% (calculated over {num_compared_frames} frames within the first {comparison_duration_sec:.2f} seconds)")
      else:
           print("Cannot calculate accuracy: Classification results within the comparison duration are all invalid (NaN).")


    else:
      print("Cannot calculate accuracy: No valid comparison duration or zero frames to compare.")


    # Optional: Add the ground truth to the plot for visual comparison
    if num_frames > 0: # Only plot if there is data
      # Create a step plot for ground truth over the full signal duration
      # We need x values that define the start and end of each segment
      gt_plot_t = np.concatenate(([t[0]], t[gt_idx_10s:gt_idx_40s], [t[gt_idx_40s-1] if gt_idx_40s > 0 else t[-1]])) # Include start, segment boundaries, and end
      gt_plot_labels = np.concatenate(([ground_truth_labels[0]], ground_truth_labels[gt_idx_10s:gt_idx_40s-1], [ground_truth_labels[gt_idx_40s-1] if gt_idx_40s > 0 else ground_truth_labels[-1]])) # Labels at boundaries + one before end
      # Adjust labels to match the 'step' plotting style
      gt_plot_t = np.insert(gt_plot_t, np.arange(1, len(gt_plot_t)), gt_plot_t[1:])
      gt_plot_labels = np.insert(gt_plot_labels, np.arange(1, len(gt_plot_labels)), gt_plot_labels[:-1])

      # Need to handle the segments carefully for step plotting over the full range
      # Create a ground truth signal array with step changes at the correct times/indices
      gt_plot_signal = np.zeros_like(t, dtype=int)
      gt_plot_signal[gt_idx_10s:gt_idx_20s] = 1
      gt_plot_signal[gt_idx_30s:gt_idx_40s] = 1 # This correctly overwrites the 20-30s segment back to 0

      # Plot the ground truth signal using step='post'
      axes[2].step(t, gt_plot_signal + 0.05, 'k--', alpha=0.7, where='post', label='Ground Truth Timing (0-40s Pattern)') # Plot slightly offset


      # Ensure the legend is updated
      handles, labels = axes[2].get_legend_handles_labels()
      # Remove duplicates if any (e.g., from initial fill_between legends)
      unique_labels = []
      unique_handles = []
      for h, l in zip(handles, labels):
          if l not in unique_labels:
              unique_labels.append(l)
              unique_handles.append(h)
      axes[2].legend(unique_handles, unique_labels, fontsize='small')

else:
    print("Error: Could not calculate accuracy. Classification results might not be available or have incorrect length, or signal is empty.")
    # Ensure ground_truth_labels exists as an empty array if num_frames is 0
    if num_frames == 0:
        ground_truth_labels = np.array([])


# <<<--- MODIFIED CODE BLOCK END --->>>

# Adjust layout and show the plots
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
plt.show()

# Final Instructions
print("\nCode execution complete. Please carefully examine the plots:")
print("1. Histogram: Helps evaluate the automatic threshold suggestion.")
print("2. Result Plot:")
print("    - Top plot: Extracted I/Q signals.")
print("    - Middle plot: Activity Index vs. the FINAL chosen threshold.")
print("    - Bottom plot: Final classification (Still/Motion) and Ground Truth comparison.")
# Repeat accuracy, handling the case where it might be NaN
accuracy_str = f"{accuracy:.2f}%" if not np.isnan(accuracy) else "Not Calculated"
print(f"=> Accuracy vs. specified pattern (10s Still/10s Motion for 0-40s): {accuracy_str}")
print("\n=> If the results or accuracy are not ideal, review the plots,")
print("   FIND the 'MANUAL THRESHOLD ADJUSTMENT SECTION' above (marked with <<<--- !!! ... !!! --->>>),")
print("   modify the 'motion_threshold = ...' line there based on the histogram and activity index plot,")
print("   and RE-RUN this cell.")
print("\nRemember to check the 'file_name', 'target_range_bin_index', 'target_antenna_index',")
print("and 'FRAME_PERIOD' variables at the beginning of the code to match your data.")
```
